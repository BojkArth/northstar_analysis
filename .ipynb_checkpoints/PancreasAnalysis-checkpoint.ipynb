```python
def cellTypeOccurrence(tsnedf,metadf):
    tsnedf['Tumor'] = metadf['Tumor']
    tsnedf['count'] = 1
    counts = pd.pivot_table(tsnedf,index='new_membership',columns='Tumor',values='count',aggfunc=np.sum)
    return counts
```
